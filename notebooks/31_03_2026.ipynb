```python
import numpy as np
import scipy.linalg as la
```

```python
from qqe.GNN.training.utils import collect_files_path, cache_root_paths
from qqe.GNN.training.datasets import build_loaders_NN
```

```python
def haar_unitary_gate(d, rng):
    a, b = rng.normal(size=(d, d)), rng.normal(size=(d, d))

    Z = a + 1j * b
    Q, R = np.linalg.qr(Z)

    Lambda = np.diag([R[i, i] / np.abs(R[i, i]) for i in range(d)])
    return np.dot(Q, Lambda)
```

```python
def eigenvalues_bins(X):
    lam = la.eigvals(X)
    phase = np.angle(lam)
    phases = np.mod(phase, 2 * np.pi)
    counts, _ = np.histogram(phases, bins=50, range=(0.0, 2 * np.pi))

    return counts
```
